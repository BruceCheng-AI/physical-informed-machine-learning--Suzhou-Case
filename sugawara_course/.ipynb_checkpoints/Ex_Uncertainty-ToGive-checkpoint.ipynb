## 2. Sugawara tank model

The Sugawara (Sugawara 1961) tank is an extension of the Linear Reservoir model which uses several (linear or non-linear) tanks. The flexibility in the selection of the number of tanks, and its response, provides a more complex representation of the phenomena, which not necessarily yields better simulation results. A 2 linear tanks Sugawara model is presented in Figure 4, however the configuration of the tanks is arbitrary and has to be defined by the modeller.

<img src='Figures/Sugawara1.png' width="400"></img>

Where P stands for precipitation and ET for evapotranspiration. S1 and S2 are the model states, represented by the water level in each of the tanks. d1 and d2 are the position of the outlets in the top tank. And k1to k4 represent the flux constants. Therefore, d and k are the model parameters to be determine during calibration. The description of the model components is presented in Table 1.

# 6. Monte Carlo simulations

## 6.1 Random number generation

Monte Carlo (MC) simulation techniques have been applied for uncertainty analysis very successfully in the last decades in hydrological sciences, and have lead the quantification of the model output uncertainty resulting from uncertain model parameters, input data or model parameters or its structure. The technique involves random sampling from the distributions of uncertain inputs and the model is run successively until a desired statistically significant distribution of outputs is obtained. The main advantage of the MC simulation based uncertainty analysis is its general applicability; however, me0thods of this type require a large number of samples (or model runs), and their applicability is sometimes limited to simple models.

<img src='Figures/Montecarlo1.png'></img>

Figure 6. Flow chart of Monte Carlo simulation to analyse parameter uncertainty.

The procedure for the application of MCS to assess the model output uncertainty for uncorrelated random variables is described above (see Figure 6). Let the model under consideration consists of input random (uncertain) variables X1,, Xn defined by their respective probability density functions (pdfs) f(x1); and output random variable Y. 


## 6.2 Sample data from a given distribution

> The following procedure assumes you have data from an input parameter such that you can first estimate your PDF. After having the PDF you calculate the CDF. Having the CDF you could random generate values for the distribution obtained.

1.	Derive cumulative distribution function (cdf) F(x) of input random variables X1 from its pdf f(x1). Note that  
2.	Generate uniform random number u1 between 0 and 1. 
3.	Determine value of the random variable x1 corresponding to u1 by inverse cdf method (Figure 7). In this method, the cdf of the random variable is equal to the generated random number i.e.
4.	Repeat the steps 1 to 3 for other input random variables X2,, Xn. This will give the values of x1,, xn as input to the model. 
5.	Run the model with the set of random variables x1,, xn as input. This gives the y1 as output of the model for first run.
6.	Repeat the steps 1 to 5 as many times as required. This gives a set of outputs y1,, yN, where N is the number of model runs.
7.	From the set of outputs y1,, yN, derive the cdf and other statistical properties (e.g. mean and standard deviation) of Y.

## 6.5 Running the montecarlo simulation
> - Proceed to run the model in the MC loop

> 1. Define the number of runs for the MC experiment

> 2. Make the random sampling of the parameters using a uniform distribution. The parameters of the distribution are assigned by the designer. In this case we use a uniform distribution


```python

mc_runs = 2000


# Make the random samples (complete set). This will be a matrix of the number of runs by the number of parameters

# First we need the range of the random generation, 
# and for sake of simplicity in this exercise we will 
# just generate it around 0.5 and 1.5 time the optimal parameter found or the parameter set provided
rand_pars = np.ones([mc_runs, len(pars)])


for i in range(len(pars)):
    x =  np.random.uniform(pars[i]*0.5,pars[i]*1.5, size=mc_runs)
    rand_pars[:,i] = x

# run the model in the MC loop
q_sim_mc = np.zeros([len(prec), mc_runs])
for i in range(mc_runs):
    q_sim = sugawara.simulate(prec=prec[:-1], 
                              evap=evap[:-1], 
                              param=rand_pars[i, :], 
                              extra_param=extra_pars)[0]
    q_sim_mc[:, i] = q_sim
    
```

### 6.6 Visualization of the montecarlo run

```python
# Visualise the results of the MC experiment. 
# Plot the hydrograph of recoded and MC results (Spaghetti plot)
plt.plot(q_sim_mc[:,0], color='#a6a6a6', label='MC')
plt.plot(q_sim_mc[:,1:], color='#a6a6a6')
plt.plot(np.array(data['Qrec']), color='k', linewidth=2, label='Rec')
plt.legend()

# Set the vertical limit of the plot
plt.ylim(0, 100)

# Display the plots
plt.show()
```

## 6.7 Running the MC experiment for convergence

As you know, the selection of the number of runs have been so far decided by the designer of the experiment. However, an additional criteria can be included, such that we ensure that the number of runs are adequate to characterise one of the moments of the PDF on the MC experiments. Traditionally, it is commonly use to characterise only the first moment of the PDF (mean), to a given value. Larger uncertainty in the input variables, and higher non-linearity of the models will result in more iterations to ensure the convergence of the MC simulation. 

As in this case the number of runs is not known from the start, we will be running the model until the convergence is lower (or equal) to the threshold we set. Threfore, it is expected to use a `while` loop to carry-out the experiment. The convergence is measured in terms of the variance of the standard deviation of the mean to a number of observations. Naturally, we should have more than 1 observation to calculate the variance.

Here we are going to deal with this approach for two indicators:
    
- Convergence of the PDF at time 256
- Convergence of the PDF of the model error (RMSE)

### 6.7.1 Code for convergence of PDF at time 256


```python
# Now we are running a montecarlo

tr = 0.3  # Threshold for the standard error. Values of 0.01 are not strange!
st_error = 10  # Initial value of standard error so we can go inside the while loop
pars_store = []  # container to store the random parameters
q_sim_mc = []  # contained to store the MC simulation results (Q mc)
st_error_store = []  # container to store the standard error values (to use for visualisation later)

# We run the loop until the standard error is lower than the threshold we set
while st_error > tr:
    
    # Make the random samples
    rand_pars = []
    for i in range(len(pars)):
        x =  np.random.uniform(pars[i]*0.5, pars[i]*1.5)
        rand_pars.append(x)
    
    pars_store.append(rand_pars)
    
    # run the model in the MC loop
    
    q_sim = sugawara.simulate(prec=prec[:-1], 
                              evap=evap[:-1], 
                              param=rand_pars, 
                              extra_param=extra_pars)[0]
    q_sim_mc.append(q_sim)

    # We slice the time series to get the results only in step 256
    tmp = np.array(q_sim_mc)[:, 256]
    
    # We need some initial iterations to be able to calculate the standard deviation
    if len(q_sim_mc) > 3:
        st_error = np.std(tmp)/np.sqrt(len(tmp))  # we calculate the standard error
        st_error_store.append(st_error)  # Store the standard error

print('Number of iterations to achieve convergence: {0}'.format(len(q_sim_mc)))

plt.plot(st_error_store)
plt.xlabel('Iterations')
plt.ylabel('Standard Error')
plt.grid()
plt.show()
```

### 6.7.2 Calculate the convergenceof the model error PDF (RMSE)

```python
# Now we are running a montecarlo

tr = 0.3  # Threshold for the standard error
st_error = 10  # Initial value of standard error so we can go inside the while loop

pars_store = []  # container to store the random parameters
st_error_store = []  # container to store the standard error values (to use for visualisation later)
rmse_store = []  # container to store the results of the model error

# We run the loop until the standard error is lower than the threshold we set
while st_error > tr:
    
    # Make the random samples
    rand_pars = []
    for i in range(len(pars)):
        x =  np.random.uniform(pars[i]*0.5, pars[i]*1.5)
        rand_pars.append(x)
    pars_store.append(rand_pars)
    
    # run the model in the MC loop
    q_sim = sugawara.simulate(prec=prec[:-1], 
                              evap=evap[:-1], 
                              param=rand_pars, 
                              extra_param=extra_pars)[0]
    
    # Calculate the RMSE and store
    rmse = np.sqrt(np.sum((q_sim - data['Qrec'])**2)/len(q_sim))
    rmse_store.append(rmse)
    
    # We need some initial iterations to be able to calculate the standard deviation
    if len(rmse_store) > 3:
        st_error = np.std(rmse_store)/np.sqrt(len(rmse_store)) # we calculate the standard error
        st_error_store.append(st_error) # we store the standard error

print('Number of iterations to achieve convergence: {0}'.format(len(rmse_store)))

```

```python
# plot of the standard errpr
plt.plot(st_error_store)
plt.xlabel('Iterations')
plt.ylabel('Standard Error')
plt.grid()
plt.show()
```
